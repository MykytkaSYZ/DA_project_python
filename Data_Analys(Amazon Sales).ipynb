# Amazon Sales Data Analysis and Visualization
In this project, I demonstrated skills in data preprocessing and visualization using Python.
I worked with an Amazon sales dataset, performing data cleaning tasks such as handling missing values, converting text columns into numerical formats, removing duplicates, and correcting inconsistent entries.
After preprocessing, I created visualizations to uncover insights: including product distribution by category, top-rated products, correlation heatmaps, average ratings by category, and products with the highest discounts.
Tools used: pandas, matplotlib, seaborn.



# Data Preprocessing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('amazon.csv')
df.head()
```

```python
df['category'].unique()
```

To simplify the analysis, I decided to shorten the category labels.
Since the original categories were hierarchical and detailed (e.g., "Electronics|HomeTheater,TV&Video|Accessories|Cables|HDMICables"), I extracted only the top-level category by splitting the string at the "|" separator and keeping the first element.

```python
df['category']=df['category'].str.split('|').str[0]
```

```python
df['category'].unique()
```

```python
df.head()
```

```python
df.info()
```

```python
#Drop duplicates by product_id so that the product is not repeated

df['product_id']=df['product_id'].drop_duplicates()
```

```python
#Remove empty values
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#To prepare the pricing data for numerical analysis, I cleaned the price-related columns by removing non-numeric characters.

df['discounted_price']=df['discounted_price'].str.replace('₹','')
df['actual_price']=df['actual_price'].str.replace('₹','')
df['discount_percentage']=df['discount_percentage'].str.replace('%','')
```

```python
df
```

```python
df = df[df['rating'] != '|']
```

```python
df['rating'].unique()
```

```python
#After removing the unwanted characters, I further cleaned the data by converting the relevant columns to the appropriate numeric types
#so that calculations and visualizations would be error-free

df['discounted_price'] = df['discounted_price'].str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace(',', '').astype(float)
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(int)
df['discount_percentage']=df['discount_percentage'].astype(int)
df['rating']=df['rating'].astype(float)
```

```python
df
```

```python
df.info()
```

```python
#Delete columns that we no longer need

df = df.drop(columns = ["user_id","review_id","img_link","product_link","user_name"])
df
```

# Data Visualization


```python
category_counts = df["category"].value_counts()
print(category_counts)
```

```python
#bar
plt.figure(figsize=(10, 8))
sns.barplot(x=category_counts.index, y=category_counts.values)

plt.title("The number of products by category")
plt.xlabel("category")
plt.ylabel("Number of products")
plt.xticks(rotation=45,fontsize=8)
plt.tight_layout()
plt.show()
```

**Observation:**  
The "Electronics" category has the highest number of products, highlighting a strong focus on tech and gadgets in this dataset.



```python

top_products = df.sort_values(by='rating', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

ax = sns.barplot(
    data=top_products,
    y="product_name",
    x="rating",
    palette="Blues_d"
)

for i in ax.containers:
    ax.bar_label(i, fmt='%.1f', label_type='edge', padding=5)

plt.title("Top 10 Products by Rating", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Product Name", fontsize=12)
plt.yticks(fontsize=10)
plt.tight_layout()

plt.show()
```

**Observation:**

The top-rated products mainly come from a variety of categories, showing that high ratings are not limited to a single type of product.


```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation matrix")
plt.show()
```

**Observation:**

There is a strong positive correlation between actual price and discounted price, while other variables show weaker relationships.


```python
category_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=category_rating, y='category', x='rating', palette='Blues_d')
plt.title("Average rating by category")
plt.xlabel("Average rating")
plt.ylabel("Сategory")
plt.tight_layout()
plt.show()
```

**Observation:**

The Office Supplies and Toys & Games categories have the highest scores, while Cars & Motorcycles has the lowest. Overall, all categories are rated high, around 4 points.

```python
top_discount = df.sort_values(by='discount_percentage', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=top_discount, y='product_name', x='discount_percentage', palette='Greens_r')
plt.title("Top 10 Products with the Biggest Discount")
plt.xlabel("Discount (%)")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='actual_price', y='rating', hue='category', alpha=0.7)
plt.title('Actual Price vs Rating by Category')
plt.xlabel('Actual Price')
plt.ylabel('Rating')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Observation:**

The relationship between price and rating is weak:
There is no clear relationship between price and rating: an expensive product does not necessarily have a higher rating, and vice versa.

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='category', y='discount_percentage')
plt.xticks(rotation=45)
plt.title('Discount percentage by Category')
plt.xlabel('Category')
plt.ylabel('Discount')
plt.tight_layout()
plt.show()
```

**Observation:**

The most generous discounts are in electronics and computers.

The most stable discounts are in household and health products.

There are almost no discounts on toys.

Office Products rarely pleases with discounts.

# Conclusion

Most products are concentrated in the Electronics category, with consistently high ratings across all price ranges.

There is no strong link between price and rating — both expensive and affordable products can have high ratings.

Electronics and Computers & Accessories offer the largest discounts, often up to 90%, while still maintaining high customer ratings. A slight positive correlation exists between discount percentage and rating, suggesting that discounts may enhance customer satisfaction.

Categories like Office Products have the highest average ratings but offer smaller discounts. Meanwhile, Home Improvement and Health & Personal Care show stable, moderate discounts, and Toys & Games rarely offer any discounts at all.

Overall, discounts vary by category, but high product ratings are consistent across the board.
